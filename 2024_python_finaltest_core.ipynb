## 함수 기초

```python
# %%
print('안녕, 현수!')
print('안녕, 수희!')
print('안녕, 지수!')

# %% 함수
def greet():
    print('안녕, 현수!')
    print('안녕, 수희!')
    print('안녕, 지수!')

# greet()
```

```python
# %%
def greet(name):
    print(f'안녕, {name}!')

greet('현수')
greet('수희')
greet('지수')
greet('모든 친구들')
```

```python
# %% 반환 값 None
def greet(name):
    print(f'안녕, {name}!')

print(greet('모든 친구들'))
```

```python
# %%
def toArea(pg):
    return pg * 3.3

print(toArea(16))

pg = 24
print(f'{pg} 평은 {toArea(pg):.3f} 제곱미터')
pg = 32
print(f'{pg} 평은 {toArea(pg):.2f} 제곱미터')
```

#### 함수도 객체

```python
# 함수 정의
def hello():
    print('안녕')

print(type(hello))
```

```python
# 정의된 함수 주소 출력
print(hello)
print(id(hello))
```

```python
# 함수 호출
print(hello())

# 함수 대입
my_hello = hello
my_hello()

id(my_hello), id(hello)
```

```python
# %% 람다 함수 정의
lambda : print('안녕')
```

```python
# 람다 함수 정의와 호출
(lambda : print('안녕'))()
```

```python
my_hello = lambda : print('안녕')
my_hello()
```

```python
print(my_hello, type(my_hello))
my_hello()
```

```python
# 람다 함수 정의
pyeong_to_ms = lambda pg: f'{pg} 평은 {pg*3.3:.2f} 제곱미터 입니다.'

# 람다 함수 호출
print(pyeong_to_ms(24))
print(pyeong_to_ms(32))
```

# 리스트 컴프리헨션

```python
# 간단한 리스트 컴프리헨션
[i for i in range(5)]
```

```python
[i for i in range(20)]
```

```python
# 연산식의 다양한 표현
print([i + 1 for i in range(5)])
print([i ** 2 for i in range(5)])
print([pow(i, 3) for i in range(5)])
```

```python
# 간단한 리스트 컴프리헨션
lst1 = [i for i in range(10)]
lst2 = [[i, i+1] for i in range(5)]
lst3 = [{i+1: str(i+1) + '월'} for i in range(5)]

print(lst1)
print(lst2)
print(lst3)
```

```python
# 반복 전체의 항목을 리스트에 삽입
print([i for i in range(6)])
# 반복 항목 중에서 짝수를 리스트에 삽입
print([i for i in range(6) if i%2 == 0])
# 반복 항목 중에서 홀수를 리스트에 삽입
print([i for i in range(6) if i%2 == 1])
```

```python
i = 1
'even' if i%2 == 0 else 'odd'
```

```python
# 리스트에 짝수면 0, 홀수면 1을 삽입
[0 if i%2 == 0 else 1 for i in range(10)]
```

```python
# 리스트에 짝수면 even, 홀수면 odd를 삽입
['even' if i%2 == 0 else 'odd' for i in range(10)]
```

```python
# 고정된 항목의 중첩된 2차원 리스트
[[1, 2] for i in range(4)] # 고정된 항목
```

```python
# 변화되는 중첩된 2차원 리스트
[[i, i+1, i+2] for i in range(3)]
```

```python
# 다양한 자료형으로 구성된 리스트의 항목
print([(i, i+1) for i in range(4)])
print([{i, i+1} for i in range(4)])
print([{i: i**2} for i in range(4)])
```
